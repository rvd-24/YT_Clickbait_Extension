```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data=pd.read_csv('/content/drive/MyDrive/Merged.csv.xls')
data.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
data=data[['Thumbnail','label']]
data
```

```python
import PIL
import urllib
from PIL import Image
urllib.request.urlretrieve(data['Thumbnail'][34], 'filename.jpg')
img=Image.open('filename.jpg')
img=img.resize((255,255))
plt.axis('off')
plt.imshow(img)
```

```python
df=pd.DataFrame(columns=['image','label'])
```

```python
for i in range(0,len(data)):
  urllib.request.urlretrieve(data['Thumbnail'][i], 'img{}.jpg'.format(i))
  if data['label'][i]==1:
    label=1
  else:
    label=0

  it={'image':'img{}.jpg'.format(i),'label':label}
  df.loc[len(df)]=it
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df['image'], df['label'] , random_state=10, train_size=0.85, shuffle=True)
```

```python
from PIL import ImageOps
def extract_features(images):
    features=[]
    for image in images:
        img=Image.open(image)
        img=img.resize((256,256))
        img=np.array(img)
        features.append(img)
    
    features=np.array(features)
    features=features.reshape(len(features),256,256,3)
    return features
```

```python
X_train=extract_features(X_train)
```

```python
X_train.shape
```

```python
X_train=X_train/255.0
```

```python
X_test=extract_features(X_test)
X_test=X_test/255.0
```

```python
X_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
```

```python
model=Sequential([
    Conv2D(32,kernel_size=(3,3),activation='relu',input_shape=(256,256,3)),
    MaxPooling2D(pool_size=(2,2)),

    Conv2D(64,kernel_size=(3,3),activation='relu'),
    MaxPooling2D(pool_size=(2,2)),

    Conv2D(128,kernel_size=(3,3),activation='relu'),
    MaxPooling2D(pool_size=(2,2)),

    Conv2D(256,kernel_size=(3,3),activation='relu'),
    MaxPooling2D(pool_size=(2,2)),

    Flatten(),
    Dense(64,activation='relu'),
    Dropout(0.3),
    Dense(1,activation='sigmoid'),
])
```

```python
model.summary()
```

```python
model.compile(loss=['binary_crossentropy'], optimizer='Adam', metrics=['accuracy'])
```

```python
X_train2 = np.asarray(X_train).astype('float32')
y_train2 = np.asarray(y_train).astype('float32')
```

```python
model.fit(X_train2,y_train2, epochs=20, batch_size=20, validation_split=0.2)
```

```python
X_test2 = np.asarray(X_test).astype('float32')
y_test2 = np.asarray(y_test).astype('float32')
```

```python
model.evaluate(X_test2,y_test2)
```

```python
pred=[]
```

```python
for i in range(0,len(X_test2)):
  p=model.predict(X_test2[i].reshape(1,256,256,3))
  p=p>0.5
  pred.append(p)
```

```python
pred=np.array(pred)
```
