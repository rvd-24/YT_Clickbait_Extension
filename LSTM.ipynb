```python
voc_size = 1500# Parameter indicating the number of words we'll put in the dictionary
VAL_SIZE = 1000  # Size of the validation set
epochs = 50  # Number of epochs to train the model for
batch = 32  # Size of the batches used in the mini-batch gradient descent
MAX_LEN = 550  # Maximum number of words in a sequence
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.Clean,df.label,test_size=0.1,random_state=37)
#samplecheck=X_test.reset_index()
```

```python
X_train.iloc[0]
```

```python
#Tokenizing the preprocessed tweets, i.e. converting the text into vectors
tk = Tokenizer(num_words=voc_size,
filters='!"#$%&()*+,-./:;<=>?@[\]^_`{"}~\t\n',split=" ")
tk.fit_on_texts(X_train)
X_train_seq = tk.texts_to_sequences(X_train)
X_test_seq = tk.texts_to_sequences(X_test)
```

```python
X_train_seq
```

```python
X_train_seq_trunc = pad_sequences(X_train_seq, maxlen=MAX_LEN)
X_test_seq_trunc = pad_sequences(X_test_seq, maxlen=MAX_LEN)
```

```python
X_train_seq_trunc
```

```python
X_train_emb, X_valid_emb, y_train_emb, y_valid_emb = train_test_split(X_train_seq_trunc, y_train, test_size=0.1, random_state=37)
```

```python
X_train_emb
```

```python
#Building the model with Embedding layer, LSTM(100 neurons) and a Dense Layer(3 neurons)
model = Sequential()
model.add(Embedding(voc_size, 32, input_length=MAX_LEN))
model.add(LSTM(64,dropout=0.2, recurrent_dropout=0.2))
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam'
                  , loss='binary_crossentropy'
                  , metrics=['accuracy'])
```

```python
print(model.summary())
```

```python
history=model.fit(X_train_seq_trunc, y_train,epochs=epochs,batch_size=batch,validation_data=(X_valid_emb, y_valid_emb))
```

```python
model.evaluate(X_test_seq_trunc,y_test)
```

```python
y_pred=model.predict(X_test_seq_trunc)
y_pred
```

```python
predict=y_pred.argmax(1)
```
